```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Render our plots inline
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
sns.set_theme(style="whitegrid")
```

```python
df_2020 = pd.read_csv("2020.csv")
df_2019 = pd.read_csv("2019.csv")
df_2018 = pd.read_csv("2018.csv")
df_2017 = pd.read_csv("2017.csv")
df_2016 = pd.read_csv("2016.csv")
df_2021 = pd.read_csv("world-happiness-report-2021.csv")
```

### Taking Average from last 5 years
year from 2020 to 2016(including 2016)

```python
averages = [0] * 149
len(averages)
```

```python
averages = [0] * 149
# print(len(averages))
averages[0] = df_2020['Ladder score'].mean()
averages[1] = df_2019['Score'].mean()
averages[2] = df_2018['Score'].mean()
averages[3] = df_2017['Happiness.Score'].mean()
averages[4]= df_2016['Happiness Score'].mean()
# averages
```

```python
df1 = df_2021[['Country name','Ladder score']]
```

```python
df1.sort_values('Ladder score',ascending=False,inplace = True)
df1.reset_index(drop=True,inplace = True)
```

```python
df1["Previous Averages(All countires)"] = averages
df1.head()
```

```python
df1.set_index("Country name",inplace = True)
```

```python
# df1
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,6))
fig.suptitle("World's Happiness Data comparison")

# Bulbasaur
sns.barplot(ax=axes[0], x=df1[:10].index, y=df1["Ladder score"][:10].values)
axes[0].set_title('The Most 10 Happy Countries in 2021')
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=45)
sns.barplot(ax=axes[1], x=[2020,2019,2018,2017,2016], y=df1["Previous Averages(All countires)"][:5].values)
axes[1].set_title('Average Ladder Score for last 5 years')
axes[1].set(xlabel='All countries')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,6))
fig.suptitle("World's Unhappy Data comparison")

# Bulbasaur
sns.barplot(ax=axes[0], x=df1[-10:].index, y=df1["Ladder score"][-10:].values)
axes[0].set_title('The Most 10 Unhappy Countries in 2021')
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=45)
sns.barplot(ax=axes[1], x=[2020,2019,2018,2017,2016], y=df1["Previous Averages(All countires)"][:5].values)
axes[1].set_title('Average Ladder Score for last 5 years')
axes[1].set(xlabel='All countries')
```
